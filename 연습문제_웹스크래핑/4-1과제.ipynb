###4-1. 인구현황 데이터 분석 (필수)

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
# font name을 알고 있다면 생략가능
#한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgun.ttf'
#font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
print(font_name)
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_name)
```

```python
data = pd.read_csv('data/인구현황.csv')
```

```python
city_pop = data.groupby('행정기관')['총인구수'].sum().reset_index()

city_pop = city_pop[city_pop['행정기관'] != '전국']
top_5 = city_pop.sort_values(by='총인구수',ascending=False).reset_index(drop=True)

top_5.reset_index(drop=True).head()
```

```python
# subplots() 함수를 사용하여 axes 객체를 생성하기
figure,axes1 = plt.subplots(nrows=1,ncols=1)
figure.set_size_inches(18,12)

# print(figure)
# print(axes1)

sns.barplot(data=city_pop.sort_values(by='총인구수',ascending=False), x='행정기관', y='총인구수',ax=axes1, hue='행정기관')
# sns.barplot(data=city_pop.sort_values(by='면적',ascending=False), x='광역시도', y='면적',ax=axes2, hue='광역시도')
```

```python
city_fam = data.groupby('행정기관')['세대당 인구'].sum().reset_index()

city_fam = city_fam[city_fam['행정기관'] != '전국']
top5 = city_fam.sort_values(by='세대당 인구',ascending=False).reset_index(drop=True)

top5.reset_index(drop=True).head()
```

```python
# subplots() 함수를 사용하여 axes 객체를 생성하기
figure,axes1 = plt.subplots(nrows=1,ncols=1)
figure.set_size_inches(18,12)

# print(figure)
# print(axes1)

sns.barplot(data=city_fam.sort_values(by='세대당 인구',ascending=False), x='행정기관', y='세대당 인구',ax=axes1, hue='행정기관')
# sns.barplot(data=city_pop.sort_values(by='면적',ascending=False), x='광역시도', y='면적',ax=axes2, hue='광역시도')
```

```python
city_mwm = data.groupby('행정기관')[' '].sum().reset_index()

city_mwm = city_mwm[city_mwm['행정기관'] != '전국']
top55 = city_mwm.sort_values(by='남여 비율',ascending=False).reset_index(drop=True)

top55.reset_index(drop=True)

```

```python
# subplots() 함수를 사용하여 axes 객체를 생성하기
figure,axes1 = plt.subplots(nrows=1,ncols=1)
figure.set_size_inches(18,12)

# print(figure)
# print(axes1)

sns.barplot(data=city_mwm.sort_values(by='남여 비율',ascending=False), x='행정기관', y='남여 비율',ax=axes1, hue='행정기관')
# sns.barplot(data=city_pop.sort_values(by='면적',ascending=False), x='광역시도', y='면적',ax=axes2, hue='광역시도')

import matplotlib.pyplot as plt

# 기본 바 차트 그린 후...
plt.axhline(y=1.0, color='red', linestyle='--', linewidth=1, label='남녀 1:1 비율')
```

```python
city_mwm = data.groupby('행정기관')['남여 비율'].mean().reset_index()

city_mwm = city_mwm[city_mwm['행정기관'] != '전국']
city_mwm.loc[city_mwm['남여 비율'] > 1, '남초여초'] = '남초'
city_mwm.loc[city_mwm['남여 비율'] < 1, '남초여초'] = '여초'
city_mwm.loc[city_mwm['남여 비율'] == 1, '남초여초'] = '동일'
top55 = city_mwm.sort_values(by='남여 비율',ascending=False).reset_index(drop=True)

top55.reset_index(drop=True)


```

```python
city_fam = data.groupby('행정기관')['세대당 인구'].mean().reset_index()

avg = city_fam['세대당 인구'].mean()
over_avg = city_fam['세대당 인구'] > avg
print(f'세대당 인구 평균보다 높은 지역: 전국 평균 세대당 인구수: {avg:.2f}')

city_fam = city_fam[city_fam['행정기관'] != '전국']
top5 = city_fam.sort_values(by='세대당 인구',ascending=False).reset_index(drop=True)

top5.reset_index(drop=True).head()
```

```python
figure, (axes1, axes2) = plt.subplots(nrows=2, ncols=1)
figure.set_size_inches(18, 12)

# 상단: 남초/여초 지역 개수 (countplot!)
pop_plot = sns.countplot(x='남초여초', data=city_mwm, ax=axes1)
pop_plot.set_title('남초 vs 여초 지역개수')

# 하단: 지역별 세대수
city_house = data.groupby('행정기관')['세대수'].sum().reset_index()
city_house = city_house[city_house['행정기관'] != '전국']
area_plot = sns.barplot(x='행정기관', y='세대수', data=city_house.sort_values(by='세대수', ascending=False), ax=axes2, hue='행정기관')
area_plot.set_title('지역별 세대수')

plt.tight_layout()
plt.show()
```
