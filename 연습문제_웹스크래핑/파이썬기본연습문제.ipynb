```python
# 1. 재귀함수로 팩토리얼
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

# 2. 반복문으로 팩토리얼
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# 테스트
print(f"5! (재귀) = {factorial_recursive(5)}")
print(f"5! (반복) = {factorial_iterative(5)}")
print(f"7! (재귀) = {factorial_recursive(7)}")
print(f"7! (반복) = {factorial_iterative(7)}")
```

```python
import csv

# 1. CSV 파일에 성적 저장
students = [
    ["이름", "점수"],
    ["김철수", "85"],
    ["이영희", "92"],
    ["박민수", "78"],
    ["최수진", "95"]
]

with open("grades.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(students)

print("학생 성적이 grades.csv에 저장되었습니다.")

# 2. CSV 파일 읽어서 평균 계산
print("\n성적 분석 결과:")
scores = []

with open("grades.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader)  # 헤더(이름, 점수) 건너뛰기
    for row in reader:
        name, score = row[0], int(row[1])
        print(f"{name}: {score}점")
        scores.append(score)

avg = sum(scores) / len(scores)
print(f"전체 평균: {avg}점")
```

```python
# 1. 텍스트 파일 생성
text = """파이썬은 배우기 쉬운 프로그래밍 언어입니다
파이썬은 강력한 프로그래밍 언어입니다
파이썬을 사용하면 다양한 프로그램을 만들 수 있습니다"""

with open("sample_text.txt", "w", encoding="utf-8") as f:
    f.write(text)

# 2. 파일 읽어서 단어 빈도 계산
word_count = {}

# 찾고 싶은 키워드 목록
keywords = ["파이썬", "프로그래밍", "언어", "배우기", "쉬운", "강력한"]

with open("sample_text.txt", "r", encoding="utf-8") as f:
    content = f.read()
    for keyword in keywords:
        count = content.count(keyword)
        if count > 0:
            word_count[keyword] = count

# 3. 빈도 높은 순으로 정렬하여 출력
print("단어 빈도 분석 결과:")
sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_words:
    print(f"{word}: {count}번")
```

```python
import json

# 1. 저장할 데이터 (딕셔너리)
data = {
    "이름": "김철수",
    "나이": 25,
    "직업": "개발자",
    "취미": ["독서", "영화감상", "코딩"],
    "주소": "서울시 강남구"
}

# 2. JSON 파일에 저장
with open("data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("데이터가 data.json에 저장되었습니다.")

# 3. JSON 파일에서 읽기
print("\nJSON에서 읽어온 데이터:")
with open("data.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)

print(f"이름: {loaded['이름']}")
print(f"나이: {loaded['나이']}")
print(f"직업: {loaded['직업']}")
print(f"취미: {loaded['취미']}")
print(f"주소: {loaded['주소']}")
```

```python
# 1. 로그 데이터 정의
logs = [
    "2025-07-20 09:00:00 - INFO - 서버가 시작되었습니다",
    "2025-07-20 09:15:00 - WARNING - 메모리 사용량이 높습니다",
    "2025-07-20 10:00:00 - INFO - 사용자 로그인 성공",
    "2025-07-20 10:30:00 - ERROR - 데이터베이스 연결 실패",
    "2025-07-20 11:00:00 - INFO - 데이터 백업 완료",
    "2025-07-20 11:45:00 - ERROR - 파일을 찾을 수 없음",
    "2025-07-20 12:00:00 - WARNING - 디스크 공간 부족"
]

# 2. 로그 파일 생성
with open("app.log", "w", encoding="utf-8") as f:
    for log in logs:
        f.write(log + "\n")

print("로그 파일이 생성되었습니다.")

# 3. 특정 레벨 로그 필터링 함수
def filter_logs(filename, level):
    filtered = []
    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            if level in line:
                filtered.append(line.strip())
    return filtered

# 4. ERROR 로그 출력
print(f"\nERROR 레벨 로그들:")
for log in filter_logs("app.log", "ERROR"):
    print(log)

# 5. WARNING 로그 출력
print(f"\nWARNING 레벨 로그들:")
for log in filter_logs("app.log", "WARNING"):
    print(log)
```

### 8. 모듈과 패키지 (3문제)

```python
import math

# 함수 정의
def circle_area(radius):
    return round(math.pi * radius ** 2, 2)

def rectangle_area(width, height):
    return width * height

def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# 바로 사용 (import 필요 없음!)
print(f"원의 넓이: {circle_area(5)}")
print(f"직사각형 넓이: {rectangle_area(10, 5)}")
print(f"팩토리얼 5! = {factorial(5)}")
print(f"최대공약수(48, 18) = {gcd(48, 18)}")
```

```python
# module_usage.py
from datetime import datetime
import random

# 1. 현재 날짜와 시간
now = datetime.now()
print(f"현재 날짜와 시간: {now.strftime('%Y-%m-%d %H:%M:%S')}")

# 2. 포맷된 날짜 (요일 포함)
days = ["월요일", "화요일", "수요일", "목요일", "금요일", "토요일", "일요일"]
day_name = days[now.weekday()]
print(f"포맷된 날짜: {now.strftime('%Y년 %m월 %d일')} {day_name}")

# 3. 임의의 정수 (1~10)
print(f"임의의 숫자: {random.randint(1, 10)}")

# 4. 임의의 실수 (1.0~5.0, 소수점 2자리)
print(f"임의의 실수: {round(random.uniform(1.0, 5.0), 2)}")

# 5. 리스트에서 임의의 요소 선택
fruits = ["사과", "바나나", "오렌지", "포도", "딸기"]
print(f"임의의 리스트 요소: {random.choice(fruits)}")

# 6. 리스트 섞기
random.shuffle(fruits)
print(f"섞인 리스트: {fruits}")
```
