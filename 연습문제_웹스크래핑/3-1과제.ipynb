### 3-1. 네이버 책 검색 API 호출하기 (필수)

```python
from dotenv import load_dotenv
import os

load_dotenv()

CLIENT_ID = os.getenv("CLIENT_ID")
print(CLIENT_ID[:3])
CLIENT_SECRET = os.getenv("CLIENT_SECRET")
print(CLIENT_SECRET[:3])
```

```python
import requests
from pprint import pprint

headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET,
}


payload = {
    'query': '파이썬',
    'display': 100,
    'sort': 'sim'
}

url = 'https://openapi.naver.com/v1/search/book.json'

# requests get(url, params, headers) 요청 
res = requests.get(url, params=payload, headers=headers)
# json() 함수로 응답 결과 가져오기
print(len(res.json()['items']))
pprint(res.json()['items'])

if res.status_code == 200:
    items_data = res.json().get('items', [])
    
    with open('C:\python\Python_WebScraping_Analysis\연습문제_웹스크래핑/book.json', 'w', encoding='utf-8') as file:
        for item in items_data:
            # 도서 API 필드에 맞게 수정 (title, link, author, publisher, description 등)
            line = f"Title: {item['title']}\nAuthor: {item['author']}\nLink: {item['link']}\n{'-'*50}\n"
            file.write(line)
    print(f"Successfully saved {len(items_data)} books to 'data/nhn_books.txt'.")
else:
    print(f"Error Code: {res.status_code}")
items_data = res.json()['items']
# 'title' , 'bloggername' , 'description' , 'bloggerlink' , 'link'

# 'data/nhnblog.txt' 파일 생성하기
```

```python
# data/blog.json 파일 생성하기 
import json
import os
import pandas as pd


with open('C:\python\Python_WebScraping_Analysis\연습문제_웹스크래핑/book.json','w', encoding='utf-8') as file:
    json.dump(items_data, file)
```
